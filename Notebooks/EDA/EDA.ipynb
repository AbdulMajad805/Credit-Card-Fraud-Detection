# Data Cleaning

```python
dataset.isnull().sum()
```

```python
dataset['V24'].mean()
```

```python
dataset['V24'].fillna(dataset['V24'].mean(), inplace=True)
```

```python
dataset['V25'].fillna(dataset['V25'].mean(), inplace=True)
dataset['V26'].fillna(dataset['V26'].mean(), inplace=True)
dataset['V27'].fillna(dataset['V27'].mean(), inplace=True)
dataset['V28'].fillna(dataset['V28'].mean(), inplace=True)
```

```python
dataset['Amount'].median()
```

```python
dataset['Amount'].fillna(dataset['Amount'].median(), inplace=True)
```

```python
dataset['Class'].mode()
```

```python
dataset['Class'].fillna(dataset['Class'].mode()[0], inplace=True)
```

```python
dataset.isnull().sum().sum()
```

```python
# Removing Duplicates
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
#For reducing skewness, we need to Scale it.
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
dataset['Amount'].skew()
```

```python
sc= RobustScaler()
```

```python
dataset['Scaled_Amount']= sc.fit_transform(dataset[['Amount']])
```

```python
dataset.head(2)
```

```python
dataset.drop(['Amount'], axis=1, inplace=True)
```

```python
dataset.head(2)
```

```python
dataset['Scaled_Amount'].skew()
```

```python
# Transforming Time into Hours
```

```python
dataset['Hour'] = dataset['Time'].apply(lambda x: np.floor(x / 3600) % 24)
```

```python
dataset.head(2)
```

```python
dataset.drop(['Time'], axis=1, inplace=True)
```
